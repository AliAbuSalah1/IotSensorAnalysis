```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('data.csv')
print(df.describe())
```

```python
df['rolling_mean'] = df['temperature'].rolling(window=5).mean()
df['rolling_std']  = df['temperature'].rolling(window=5).std()

last_mean = df['rolling_mean'].iloc[-1]
last_std  = df['rolling_std'].iloc[-1]

future_ts    = list(range(26, 36))
future_mean  = [last_mean] * 10
future_upper = [last_mean + last_std] * 10
future_lower = [last_mean - last_std] * 10

fig, axes = plt.subplots(2, 1, figsize=(12, 9))

axes[0].plot(df['timestamp'], df['temperature'], 'o', color='coral', alpha=0.6, label='Actual')
axes[0].plot(df['timestamp'], df['rolling_mean'], '-', color='purple', linewidth=2, label='Rolling mean')
axes[0].plot(future_ts, future_mean, '--', color='green', linewidth=2, label='Forecast')
axes[0].fill_between(future_ts, future_lower, future_upper, color='green', alpha=0.15, label='Confidence band')
axes[0].axvline(x=25, color='gray', linestyle=':')
axes[0].set_ylabel('Temperature C')
axes[0].set_title('Rolling Mean Forecast')
axes[0].legend()

axes[1].hist(df['temperature'], bins=8, color='steelblue', edgecolor='white', alpha=0.8)
axes[1].axvline(df['temperature'].mean(), color='coral', linestyle='--', label=f"Mean: {df['temperature'].mean():.1f}C")
axes[1].set_xlabel('Temperature C')
axes[1].set_ylabel('Count')
axes[1].set_title('Temperature Distribution')
axes[1].legend()

plt.tight_layout()
plt.show()

print(f'التوقع القادم: {last_mean:.2f}C +/- {last_std:.2f}')
```
